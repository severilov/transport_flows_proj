```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from glob import glob
```

```python
def read_all_iterations(paths):
    all_iter_results = []
    for path in paths:
        iter_result = np.loadtxt(path)
        all_iter_results.append(iter_result)
    return all_iter_results

results_path = './results/multi'


sinkhorn = 'accelerated'
flows_paths = sorted(glob(f'{results_path}/flows/{sinkhorn}*'))
times_paths = sorted(glob(f'{results_path}/times/{sinkhorn}*'))
corr_matrix_paths = sorted(glob(f'{results_path}/corr_matrix/{sinkhorn}*'))
inverse_func_paths = sorted(glob(f'{results_path}/inverse_func/{sinkhorn}*'))
subg_paths = sorted(glob(f'{results_path}/subg/{sinkhorn}*'))

all_flows = read_all_iterations(flows_paths)
all_times = read_all_iterations(times_paths)
all_corr_matrix = read_all_iterations(corr_matrix_paths)
all_inverse_func = read_all_iterations(inverse_func_paths)
all_subg = read_all_iterations(subg_paths)

sinkhorn = 'base'
flows_paths = sorted(glob(f'{results_path}/flows/{sinkhorn}*'))
times_paths = sorted(glob(f'{results_path}/times/{sinkhorn}*'))
corr_matrix_paths = sorted(glob(f'{results_path}/corr_matrix/{sinkhorn}*'))
inverse_func_paths = sorted(glob(f'{results_path}/inverse_func/{sinkhorn}*'))
subg_paths = sorted(glob(f'{results_path}/subg/{sinkhorn}*'))

all_flows_base = read_all_iterations(flows_paths)
all_times_base = read_all_iterations(times_paths)
all_corr_matrix_base = read_all_iterations(corr_matrix_paths)
all_inverse_func_base = read_all_iterations(inverse_func_paths)
all_subg_base = read_all_iterations(subg_paths)
```

```python
print('='*40 + 'FLOWS' + '='*40)

for i, flows in enumerate(all_flows):
    print('-'*20 + str(i) + '-'*20)
    plt.plot(flows)
    plt.show()
```

```python
print('='*40 + 'TIMES' + '='*40)

for i, times in enumerate(all_times):
    print('-'*20 + str(i) + '-'*20)
    plt.plot(times)
    plt.show()
```

```python
print('='*40 + 'CORR MATRIX' + '='*40)

for i, corr_matrix in enumerate(all_corr_matrix):
    print('-'*20 + str(i) + '-'*20)
    plt.plot(corr_matrix)
    plt.show()
```
